#### Pie

```python
x = np.array([1, 2, 3, 4])
plt.pie(x)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.  
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
mylables = ['Apple', 'Banana', 'Cherry', 'orange']
plt.pie(x, labels=mylables, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
mylables = ['Apple', 'Banana', 'Cherry', 'orange']
explode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Apple')
plt.pie(x, labels=mylables, autopct='%1.1f%%', startangle=140, explode=explode)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
colors = ['red', 'green', 'blue', 'orange']
plt.pie(x, labels=mylables, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
mylables = ['Apple', 'Banana', 'Cherry', 'orange']
myexplode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Apple')
plt.pie(x, labels=mylables, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
mylables = ['Apple', 'Banana', 'Cherry', 'orange']
myexplode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Apple')
colors = np.array(['red', 'green', 'blue', 'orange'])
plt.pie(x, labels=mylables, autopct='%1.1f%%', startangle=140, explode=myexplode, colors=colors, shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
mylables = ['Apple', 'Banana', 'Cherry', 'orange']
myexplode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Apple')
colors = np.array(['pink', 'navy', 'gold', 'silver'])
plt.pie(x, labels=mylables, autopct='%1.1f%%', startangle=140, explode=myexplode, colors=colors, shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
mylables = ['Apple', 'Banana', 'Cherry', 'orange']
myexplode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Apple')
colors = np.array(['red', 'yellow', 'darkred', 'orange'])
plt.pie(x, labels=mylables, autopct='%1.1f%%', startangle=140, explode=myexplode, colors=colors, shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
mylables = ['Apple', 'Banana', 'Cherry', 'orange']
myexplode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Apple')
colors = np.array(['red', 'green', 'blue', 'orange'])
plt.legend(title='Fruits', loc='upper right')
plt.pie(x, labels=mylables, autopct='%1.1f%%', startangle=140, explode=myexplode, colors=colors, shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
