```python
#%pip install seaborn

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = sns.load_dataset("tips")
data
```

```python
sns.scatterplot(data=data, x="total_bill", y="tip")
plt.show()
```

```python
def new_func():
    sns.scatterplot(data=data, x="total_bill", y="tip")
    plt.xlabel("Total Bill")
    plt.ylabel("Tip")
    plt.title("Scatterplot of Total Bill vs Tip")
    plt.show()

new_func()
```

```python
data = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=data)
plt.title("Barplot of Total Bill by Day")
plt.show()
```

```python
df = sns.load_dataset("flights")
sns.lineplot(x="year", y="passengers", data=df)
plt.title("Year vs Passengers")
plt.show()
```

```python
df = sns.load_dataset("tips")
sns.histplot(df["total_bill"], kde=True, bins=20)
plt.title("Distribution of Total Bill")
plt.show()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'time', style = 'smoker', kind = 'scatter', col = "day", data= data )
```

```python
sns.relplot(y = 'total_bill', x = 'tip', data= data)
plt.title("this is relplot")
plt.xlabel("Total_Bill", fontsize = 20, color = "black")
plt.ylabel("tip", fontsize = 20, color = "red")
plt.show()
```

### Catplot

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'swarm', data = data)
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'strip', data = data)
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = data)
```

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=data)
plt.title("Swarm Plot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=data)
plt.title("boxPlot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='bar', data=data)
plt.title("bar Plot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=data)
plt.title("violin Plot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='boxen', data=data)
plt.title("Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```
