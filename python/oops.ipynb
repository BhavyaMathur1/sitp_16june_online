```python
# 1. Single Inheritance
class Parent:
    def parent_method(self):
        print("This is a method from the Parent class.")

        class Child(Parent):  # Child class inherits from Parent
           def child_method(self):
             print("hello from the Child class.")


             Child_obj = Child()  # create an instance of Child

             Child_obj.parent_method()  # calling method from Parent class
```

```python
# 2. Multiple Inheritance
class Parent1:
    father_name = "John"
class Parent2:
    mother_name = "Jane"
class Child(Parent1, Parent2):  # Child inherits from both Parent1 and Parent2
    def display_names(self):
        print(f"Father: {self.father_name}, Mother: {self.mother_name}")    
```

```python
# 3. Multilevel Inheritance
class Grandparent:
    def __init__(self):
        self.grandfather = "Grandpa Joe"

class Parent(Grandparent):  # Parent inherits from Grandparent
    def __init__(self):
        super().__init__()
        self.father = "Father Mike"

class Child(Parent):  # Child inherits from Parent
    def __init__(self):
        super().__init__()

obj = Child()  # create an instance of Child
obj.grandfather  # This will now work without error
obj.father  
```

```python


class Parent:
    pass
class Child1(Parent):
    pass
class Child2(Parent):
    pass

# Hierarchical Inheritance: Child1 and Child2 inherit from Parent

# Hybrid Inheritance Example (valid MRO)
class Family(Child1, Child2):  # Family inherits from both Child1 and Child2
    pass

fam_obj = Family()  # create an instance of Family
```

```python
#print()
#len()

class Animal:
    def sound(self):
        pass 

class Dog(Animal):
    def sound(self):
        print("Woof! Woof!")  # Dog class overrides the sound method
     
class Cat(Animal):
    def sound(self):
        print("Meow! Meow!")  # Cat class overrides the sound method


dog_obj = Dog()  # create an instance of Dog
dog_obj.sound()  # calling the overridden method

cat_obj = Cat()  # create an instance of Cat
cat_obj.sound()  # calling the overridden method
```
